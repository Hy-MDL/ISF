```python
import pandas as pd
import numpy as np

# 데이터 로드
df = pd.read_csv('/home/hyeonmin/ISF/timegan-pytorch-main/data/energy_numeric.csv', index_col=False)

# zero padding 추가 (앞에 23개의 0을 추가)
padding_size = 23
padding_df = pd.DataFrame(np.zeros((padding_size, df.shape[1])), columns=df.columns)
df_padded = pd.concat([padding_df, df])

# sliding window 크기
window_size = 24
df_padded.drop(columns=['Unnamed: 0'], inplace=True)

# 벡터화된 슬라이딩 윈도우 생성
n_samples = len(df_padded) - window_size + 1
n_features = df_padded.shape[1]

# numpy 배열로 변환하여 빠른 연산 수행
data_array = df_padded.values
windows = np.lib.stride_tricks.sliding_window_view(data_array, (window_size, n_features))
windows = windows.reshape(-1, window_size, n_features)

# 결과를 데이터프레임으로 변환
result_df = pd.DataFrame(
    windows.reshape(-1, n_features),
    columns=df_padded.columns
)

# 인덱스 추가
result_df['Idx'] = np.repeat(np.arange(n_samples), window_size)

# 결과 확인
print(f"추가된 zero padding 크기: {padding_size}")
print(f"전체 데이터 크기 (padding 포함): {len(df_padded)}")
print(f"생성된 윈도우 수: {n_samples}")
print("\n처음 10개의 데이터:")
print(result_df.head(10))
```
